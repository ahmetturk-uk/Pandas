```python
# Librarires
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#Ignore the warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
netflix = pd.read_csv('/content/netflix_titles.csv')
netflix
```

```python
netflix.info()
```

```python
#we want to read all columns and check their name to avoid errors during out work
netflix = pd.DataFrame(netflix)
netflix.head()
column_list = list(netflix.columns)
column_list
```

```python
# Drop column that is not needed in the analysis

netflix.drop(columns='duration', inplace=True)
netflix.tail()
```

```python
#checking for any N/A values
netflix.isna().sum()
```

```python
# Dropping the N/A's in the below columns
netflix = netflix.dropna(subset=['director', 'cast', 'country', 'date_added', 'rating'])
```

```python
# Checking for any N/A values again and if they been dropped
netflix.isna().sum()
```

```python
# Chech the column for the duplicates. We will check for duplicated with 'colum-name'

netflix.duplicated('director').sum()
```

```python
netflix.drop_duplicates(subset= 'director', inplace=True)
netflix.duplicated('director').sum()
```

```python
netflix.duplicated('director').sum()
```

```python
netflix.describe()
```

#Visualisation


```python
sns.set_style('darkgrid')
plt.figure(figsize=(6,4))
plt.pie(netflix['release_year'].value_counts(), labels=netflix['release_year'].value_counts().index, autopct='%1.1f%%', startangle=90)
plt.title('Release Year')
plt.show()
```
